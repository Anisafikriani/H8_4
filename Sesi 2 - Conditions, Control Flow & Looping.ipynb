# Conditional
- equals : a = b
- not equals : a != b
- less than : a < b
- less than or equal to : a <= b
- greater than : a > b
- greater than or equal to : a >= b
- these conditions can be used in several ways, most commonly in "if statements" and loops

## Introduction to the if statement


```python
# CATATAN

if <expr> :
    <statement>
#<expr> : is an expression evaluated in boolean context
#<statement> : is a valid python statement, which must be indented

Note :
- jika <expr> benar (mengevaluasi ke nilai "truthy"), 
  maka <statement> akan dieksekusi.
- jika <expr> salah, maka <statement> dilewati dan tidak dieksekusi. 
```

```python
x = 0
y = 5

if x < y :
    print ('yes') # Truthy

if y < x :
    print ('yes') #Falsy
    
if x :
    print ('yes') #Falsy
    
if y :
    print ('yes') #Truthy
    
if 'aul' in 'grault' :
    print ('yes') #Truthy
    
if 'quux' in ['foo', 'bar', 'baz'] :
    print ('yes') #Falsy
```

## Grouping Statements : Indentation an Blocks

```python
if <expr> :
    <statement>
    <statement>
    ....
    <statement>
    <folowing_statement>

Note : semua pernyataan pada tingkat indentasi yang cocok (baris 2-5)
    dianggap sebagai bagian dari blok yang sama. 
    seluruh blok dijadikan jika <expr> benar, atau dilewati 
    jika <expr> salah. apapun itu, eksekusi dilanjutkan 
    dengan <foloeing_statement> (baris 6) sesudahnya.
    
```

```python
if 'foo' in ['bar', 'baz', 'qux'] :
    print ('Expression was true')
    print ('Excuting statement in suite')
    print ('...')
    print ('Done.')
    
print ('After conditional')
```

```python
if 'foo' in ['foo', 'bar', 'baz']:
    print('Outer condition is true')
    
if 10 > 20 :
    print('Inner condition 1')
    
print ('Between inner condition')

if 10 < 20 :
    print ('Inner condition 2')
    
print ('End of outer condition')
print ('After outer condition')
```

## The Else and Elif Clauses

```python
x = 20

if x < 50 :
    print('(first suite)')
    print ('x is small')
else :
    print ('(second suite)')
    print ('x is large')
```

```python
hargaBuku = 20000
hargaMajalah = 5000
uang = 2000

if uang > hargaBuku :
    print ('beli buku')
else :
    print ('uang tidak cukup')
```

```python
#Penjelasan Penggunaan

if <expr> :
    <statement(s)>
elif <expr> :
    <statement  (s)>
elif <expr> :
    <statement(s)>
    ...
else :
    <statement(s)>
Note : Jumlah klausa elif dapat ditentukan berapapun. klausa else bersifat opsional. Jika ada, hanya boleh ada satu. dan harus ditentukan diakhir
```

```python
hargaBuku = 20000
hargaMajalah = 5000
uang = 2000

if uang > hargaBuku :
    print ('beli buku')
elif uang > hargaMajalah :
    print ('beli majalah')
else :
    print ('uang tidak cukup')
```

### Python "while" Loops

```python
n = 5 
while n > 0 :
    n -=1 # tanda - menandakan urutan kebalikan
    print (n)
```

```python
i = 1
while i < 6 :
    print (i)
    i +=1 # tanda +menandakan urutan dari awal
```

### The python break and continue Statements

```python
n = 5
while n > 0 :
    n -=1
    if n == 2 :
        break # break statement
    print (n)
print ('Loop ended')
```

```python
n = 5
while n > 0 :
    n -=1
    if n == 2 :
       continue
    print (n)
print ('Loop ended')
```

### The Else Clauses

```python
n = 5
while n > 0 :
    n -=1
    print (n)
else :
    print ('Loop done.')
```

```python
n = 5
while n > 0 :
    n -=1
    print (n)
    if n == 2 :
        break
else :
    print ('Loop done.')
```

## Altering for Loop Behavior

### The break and continue Statements

```python
for i in ['foo', 'bar', 'baz', ' qux']:
    if 'b' in i:
        break
        print(i)
```

```python
for i in ['foo', 'bar', 'baz', ' qux']:
    if 'b' in i:
        continue
        print(i)
```

### The Else Clause

```python
for i in ['foo', 'bar', 'baz', 'qux']:
    print(i)
else :
    print ('Done.') #Will execute
```

```python
for i in ['foo', 'bar', 'baz', 'qux']:
    if i == 'bar' :
        break
        print (i)
else :
    print ('Done.') #Will not execute
```
